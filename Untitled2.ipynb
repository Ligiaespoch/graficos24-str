Primer Grafico

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#datos de ejemplo
datos =pd.DataFrame({
    "Tiempo":[15, 30, 45,60],
    "Distancia": [3, 6, 9, 12]
})
datos
```

```python
#estilo general
sns.set(style="whitegrid")
```

```python
# Crear el gráfico
plt.figure(figsize=(8,5))
ax = sns.barplot(
    x="Tiempo",
    y="Distancia",
    data=datos,
    palette="Set2",
    order=datos.sort_values("Tiempo", ascending=False)["Tiempo"]
)

# Títulos y etiquetas
plt.title("Distribución por categoría", fontsize=16)
plt.xlabel("Tiempo")
plt.ylabel("Distancia")
plt.tight_layout()

# Añadir etiqueta
for i, v in enumerate(datos.sort_values("Tiempo", ascending=False)["Distancia"]):
    ax.text(i, v + 1, str(v), ha="center", fontweight="bold")

plt.show()
```

Segundo Grafico

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#datos de grafico
datos =pd.DataFrame({
    "Edad":[1,3,6,9, 12,18,24,36],
    "horas de sueño": [16,15,14.10,14, 13.50, 13.40, 13.30,12]
})
datos
```

```python
#estilo general
sns.set(style="whitegrid")
```

```python
# Crear el gráfico
plt.figure(figsize=(8,5))
ax = sns.barplot(
    x="Edad",
    y="horas de sueño",
    data=datos,
    palette="Set2",
    order=datos.sort_values("Edad", ascending=False)["Edad"]
)

# Títulos y etiquetas
plt.title("Distribución por categoría", fontsize=16)
plt.xlabel("Edad")
plt.ylabel("horas de sueño")
plt.tight_layout()

# Añadir etiqueta
for i, v in enumerate(datos.sort_values("Edad", ascending=False)["horas de sueño"]):
    ax.text(i, v + 0.2, str(v), ha="center", fontweight="bold")

plt.show()
```

Tercer Grafico

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#datos de grafico
datos =pd.DataFrame({
    "Velocidad ":[20,22,24, 26,28, 30,32, 34,36, 38,40, 42],
    "Impurezas ": [8.5, 9.5, 12, 10.5, 13.5, 15, 13.5, 15, 16.5, 16.5, 19, 18.5]
})
datos
```

```python
#estilo general
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
ax = sns.scatterplot(
    x="Velocidad ",
    y="Impurezas ",
    data=datos,
    color="green",
    s=100
)

# Títulos y etiquetas
plt.title("Relación entre Velocidad e Impurezas", fontsize=16)
plt.xlabel("Velocidad")
plt.ylabel("impurezas")
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#datos de grafico
datos =pd.DataFrame({
    "Año ":[1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009],
    "Ventas ": [15, 14, 13.5, 13, 12, 12.5, 13, 12.5, 11, 9,6 ]
})
datos
```

```python
#estilo general
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
ax = sns.scatterplot(
    x="Año ",
    y="Ventas ",
    data=datos,
    color="red",
    s=100
)

# Títulos y etiquetas
plt.title("Ventas industria musical discograficas", fontsize=16)
plt.xlabel("Año (1999-2009)")
plt.ylabel("Ventas (en billones)")
plt.tight_layout()

plt.show()
```
